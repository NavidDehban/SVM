```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import itertools
from sklearn import datasets, metrics, model_selection, svm
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
```

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['q5.csv']))
```

```python
import numpy as np
features = df.columns[0:8]
features
```

```python
y = np.array(df['target_class'])
x = np.array(df[features])
x = x.tolist()
y = y.tolist()
print(x)
print(y)
```

```python
print(len(x))
print(len(y))
```

```python
scaler = preprocessing.StandardScaler().fit(x)
x = scaler.transform(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.8,random_state=0)
```

```python
print(x_test)
```

```python
clf = svm.SVC(C=100, kernel='sigmoid')
clf = clf.fit(x_train, y_train)
acc = clf.score(x_test,y_test)
print('accuracy:',acc)
```

```python
parameters = {'kernel':['rbf'], 'C': [1, 10, 100, 500], 'gamma': [0.1, 0.3, 0.5, 0.7, 0.9]}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(x_train, y_train)
print(clf.best_params_)
acc = clf.best_estimator_.score(x_test,y_test)
print('accuracy:',acc)
```

```python
parameters = {'kernel':['linear'], 'C':  [1, 10, 100, 1000]}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(x_train, y_train)
print(clf.best_params_)
acc = clf.best_estimator_.score(x_test,y_test)
print('accuracy:',acc)
```

```python
parameters = {'kernel':['poly'], 'C': [1, 10, 100, 500], 'gamma': [0.01, 0.03, 0.05], 'degree': [2, 3, 4]}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(x_train, y_train)
print(clf.best_params_)
acc = clf.best_estimator_.score(x_test,y_test)
print('accuracy:',acc)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
clf1 = svm.SVC(C=100, kernel='linear')
clf2 = svm.SVC(C=1, kernel='linear')
clf1.fit(x_train, y_train)
clf2.fit(x_train, y_train)
acc1 = clf1.score(x_test,y_test)
acc2 = clf2.score(x_test,y_test)
y_pre1 = clf1.predict(x_test)
conf1 = confusion_matrix(y_pre1,y_test)
y_pre2 = clf2.predict(x_test)
conf2 = confusion_matrix(y_pre2,y_test)
```

```python
print('accuracy of model 1:', acc1)
print('accuracy of model 2:', acc2)
print('confusion matrix of model 1:')
print(conf1)
print('confusion matrix of model 2:')
print(conf2)
```
